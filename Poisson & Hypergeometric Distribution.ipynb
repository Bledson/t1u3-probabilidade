# 1. Distribuição Poisson

```python
import matplotlib.pyplot as plt
from scipy import linspace
from scipy.stats import binom, poisson, hypergeom
```

## PMF

```python
# lambda = 1

outcome_counts = linspace(0, 10, 11)
dist = poisson.pmf(outcome_counts, 1)

plt.bar(outcome_counts, dist, label='lambda = 1')
plt.legend()

plt.show()
```

```python
# lambda = 4

dist = poisson.pmf(outcome_counts, 4)

plt.bar(outcome_counts, dist, label='lambda = 4')
plt.legend()

plt.show()
```

```python
# lambda = 10

dist = poisson.pmf(outcome_counts, 10)

plt.bar(outcome_counts, dist, label='lambda = 10')
plt.legend()

plt.show()
```

## CDF

```python
dist = poisson.cdf(outcome_counts, 1)

plt.bar(outcome_counts, dist, label='lambda = 1')
plt.legend()

plt.show()
```

```python
dist = poisson.cdf(outcome_counts, 4)

plt.bar(outcome_counts, dist, label='lambda = 4')
plt.legend()

plt.show()
```

```python
dist = poisson.cdf(outcome_counts, 10)

plt.bar(outcome_counts, dist, label='lambda = 10')
plt.legend()

plt.show()
```

# 2. Aproximação Binomial PMF pela Poisson PMF

```python
dist_binom = binom.pmf(outcome_counts, 20, .05)
dist_poisson = poisson.pmf(outcome_counts, 1)

plt.plot(outcome_counts, dist_binom, 'ro', label='X~Binom(20, .05)')
plt.plot(outcome_counts, dist_poisson, 'bo', label='Y~Poisson(20 * .05)')
plt.legend()

plt.show()
```

# 3. Distribuição Hipergeométrica

## PMF

```python
outcome_counts = linspace(0, 20, 21)
rv = hypergeom(100, 5, 4)
dist = rv.pmf(outcome_counts)

plt.plot(outcome_counts, dist, 'bo', label='X~Hypergeom(2, 100, 5, 4)')
plt.legend()

plt.show()
```

## CDF

```python
dist = rv.cdf(outcome_counts)

plt.plot(outcome_counts, dist, 'bo', label='X~Hypergeom(2, 100, 5, 4)')
plt.legend()

plt.show()
```
